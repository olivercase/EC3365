```python
year = [2000,2001,2002,2003,2004,2005,2006,2007,2008,2009]
prod1 = [100,100,108,123,112,115,125,126,144,145]
prod2 = [103,111,99,123,134,135,130,130,143,143]

plt.plot(year, prod1, label='Country 1')
plt.plot(year, prod2, label='Country 2')
plt.xlabel('Year')
plt.ylabel('Productivity')
plt.legend(loc="upper left")
plt.xticks(range(2000,2010), fontsize=8)
plt.text(2005,136, 'Country 2', color='orange')
plt.show()
```

  # Scatter plot

```python
n=1000
xa = np.random.normal(1,1,(n,1))
xb = np.random.normal(2,1,(n,1))
noise = np.random.normal(0,1,(n,1))


ya = 1 + 1 * xa  + noise
yb = -1 + 1 * xb  + noise


plt.scatter(xa,ya, color='red', label='A', alpha=0.2)
plt.scatter(xb,yb, color='blue', label='B', alpha=0.2)
plt.legend()
plt.show()
```

```python
country = ['USA', 'UK', 'Spain', 'Sweden', 'Botswana', 'China', 'Madagascar', 'Poland']
gdp = [61355, 46878, 40225, 52349, 14692, 15134, 1557, 31739]
hdi = [0.713, 0.776, 0.736, 0.802, 0.545, 0.647, 0.385, 0.760]

data = {'Country': country, 'GDP' : gdp, 'HDI': hdi}
df= pd.DataFrame(data)
df.head(10)
```

```python
countries_to_show = ['Poland', 'USA', 'China']

plt.scatter(hdi, gdp, color='grey', alpha=0.5)
for name in countries_to_show:
  position = country.index(name)
  plt.text(hdi[position], gdp[position] + 1000, name, ha='center', va='bottom',fontsize=8)
  plt.scatter(hdi[position], gdp[position], color='red', alpha=0.5)

plt.xlim(0,1)
plt.show()
```

#Histogram


```python
xa = np.random.normal(0,1,1000)
xb = np.random.normal(2,1,1000)
plt.hist(xa, bins=50, alpha=0.5, color='blue', edgecolor='blue',linewidth=0.3, label='Group A')
plt.hist(xb, bins=50, alpha=0.5, color='red', edgecolor='red', linewidth=0.3, label = 'Group B')
plt.legend(bbox_to_anchor=(1.25, 1))
plt.show()
```

#SeaBourn

```python


na = 100
nb = 200

# Creating x and y data for both groups
xa = np.random.normal(0, 1, na)
xb = np.random.normal(1, 1, nb)

ya = 1 + 1 * xa + np.random.normal(0, 1, na)
yb = 1 + 0.5 * xb + np.random.normal(0, 1, nb)

# Creating the 'group' array
group = ['A'] * na + ['B'] * nb

# Concatenating x and y data
x = list(xa) + list(xb)
y = list(ya) + list(yb)  # This line was missing in your code

# Creating the DataFrame
data = {'Group': group, 'X': x, 'Y': y}
df = pd.DataFrame(data)
print(df)
```

```python


plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='Group', y='X', palette=['skyblue', 'orange'], dodge=True)
plt.show()
```
